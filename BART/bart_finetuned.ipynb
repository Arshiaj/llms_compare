```python
from datasets import load_dataset
from transformers import AutoTokenizer, pipeline
import pandas as pd
from sklearn.metrics import classification_report
```

```python
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
def bart_classifier(dataset, labels):
    data = load_dataset('glue', dataset)['validation']
    df = pd.DataFrame(data)

    pred = []
    for i in range(len(data)):
        d = {}
        sequence_to_classify = data[i]['sentence']
        c = classifier(sequence_to_classify, labels)
        d[c['labels'][0]] = c['scores'][0]
        d[c['labels'][1]] = c['scores'][1]
        if d[labels[0]] >= 0.5:
            pred.append(1)
        else:
            pred.append(0)

    return classification_report(list(df['label']), pred)
```

```python
cr_sst2 = bart_classifier('sst2', ['positive', 'negative'])
print(cr_sst2)
```

```python
cr_cola = bart_classifier('cola', ['correct', 'incorrect'])
print(cr_cola)
```
